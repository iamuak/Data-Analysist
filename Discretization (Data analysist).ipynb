```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import KBinsDiscretizer
```

```python
df = pd.read_csv('winequality-red.csv')
df.head(15)
```

```python
sns.histplot(df['alcohol'],kde = True)
print('Orignal distibution')
```

```python
alcohol_discretizer = KBinsDiscretizer(n_bins=5, encode='ordinal', strategy='uniform')
df['alcohol_bin'] = alcohol_discretizer.fit_transform(df[['alcohol']])
df.head(15)
```

```python
sns.histplot(df['alcohol_bin'],kde = True)
print('discretized distribution')
```

```python
#Answer2(A)
residualsugar=KBinsDiscretizer(n_bins=4,encode='ordinal',strategy='uniform')
df['residual_sugar_bin']=residualsugar.fit_transform(df[['residual sugar']])
df.head(15)
```

```python
print("Histogram for the uniform strategy")
sns.histplot(df["residual_sugar_bin"],kde=True)
```

```python
#Answer2(B)
residualsugar=KBinsDiscretizer(n_bins=4,encode='ordinal',strategy='quantile')
df['residual_sugar_bin']=residualsugar.fit_transform(df[['residual sugar']])
df.head(15)
```

```python
print("Histogram for the uniform strategy")
sns.histplot(df["residual_sugar_bin"],kde=True)
```

```python
df = pd.read_csv('winequality-red.csv')
bins = [8, 9, 10, 11, 12, 13, 14, 15]
labels = ['very low', 'low', 'medium-low', 'medium', 'medium-high', 'high', 'very high']
df['alcohol_level'] = pd.cut(df['alcohol'], bins=bins)
plt.figure(figsize=(10, 6))
sns.boxplot(x='alcohol_level', y='quality', data=df, palette='Set2')
plt.title('Quality Scores by Alcohol Level')
plt.xlabel('Alcohol Level')
plt.ylabel('Quality Score')
plt.show()
```
