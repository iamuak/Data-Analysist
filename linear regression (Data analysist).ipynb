```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
# Task 1
df = pd.read_csv('canada_per.csv')
df
```

```python
# graph is just for practice
print('year_data:',df.year.head())
print('income_data:',df.income.head())
plt.xlabel('year')
plt.ylabel('income')
plt.scatter(df.year, df.income, color='purple', marker='.')
plt.show()
```

```python
model = linear_model.LinearRegression()
model.fit(df[['year']],df.income)
```

```python
model.predict([[2020]])
```

```python
# Task2
df = pd.read_csv('hiring.csv')
df.rename(columns={'salary($)': 'salary'}, inplace=True)
df
```

```python
# test_score = df['test_score(out of 10)'].median()
# df['test_score(out of 10)'].fillna(test_score, inplace=True)
# df
df['test_score(out of 10)'].fillna(0, inplace=True)
df
```

```python
df['experience'].fillna(0, inplace=True)
df
```

```python
experience_mapping = {
    'zero': 0,'two': 2, 'three': 3,
    'five': 5,'seven': 7, 'nine': 9,
    'ten': 10, 'eleven': 11
}
df['experience'] = df['experience'].replace(experience_mapping)
df
```

```python
model = linear_model.LinearRegression()
model.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']],df.salary)
```

```python
model.predict([[5,8.0,9]])
```
