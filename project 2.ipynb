```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = sm.datasets.get_rdataset("cars").data
df
```

```python
# make a graph of data set
plt.figure(figsize=(10,5))
plt.scatter(df['speed'], df['dist'], color='orange', marker='o')
plt.title('Scatter plot of Speed vs Distance')
plt.xlabel('Speed')
plt.ylabel('Distance')
plt.show()
```

```python
print('the speed increase, the stopping distance also increase, but not in a perfect straight line the relationship biuld to be nonlinear. at higher speed the stopping distance grow much faster showing a curved upward trend  that distance increase more rapidly with speed.')
```

```python
X = df[['speed']].values
y = df['dist'].values
```

```python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
```

```python
model = LinearRegression()
model.fit(X_poly, y)
y_pred = model.predict(X_poly)
```

```python
plt.scatter(X, y, color='blue', label="Original data")
plt.plot(X, y, color='red', label="polynomial Regression Line")
plt.xlabel("Speed")
plt.ylabel("Distance")
plt.title("Polynomial Regression")
plt.legend()
plt.show()
```

```python
print('the red curve click the nonlinear relationship between speed and stopping distance. the speed increase the stopping distance grow rapidly showing a clear upward curving trend.')
```

```python
# a graph with linear regression
X = df[['speed']]
y = df['dist']
model = LinearRegression()
model.fit(X, y)
X_line = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_line = model.predict(X_line)
plt.figure(figsize=(8,5))
plt.scatter(X, y, color='purple', label="Data points")
plt.plot(X_line, y_line, color='red', label="Linear Regression Line")
plt.title('Speed vs Distance with Linear Regression')
plt.xlabel('Speed')
plt.ylabel('Distance')
plt.legend()
plt.show()
```

```python
y_pred_linear = model.predict(X)
mse_linear = mean_squared_error(y, y_pred_linear)
r2_linear = r2_score(y, y_pred_linear)
mse_poly = mean_squared_error(y.values, y_pred)
r2_poly = r2_score(y.values, y_pred)
print(f"Linear Regression - MSE: {mse_linear:.2f}, R²: {r2_linear:.4f}")
print(f"Polynomial Regression - MSE: {mse_poly:.2f}, R²: {r2_poly:.4f}")
```
