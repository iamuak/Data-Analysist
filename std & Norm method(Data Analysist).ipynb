```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from scipy.stats import zscore
```

```python
#Answer1
data = {
    'Age':[25,22,25,28,19,18,30,16,50,40,35,80,25,20,30,95,62,51,21,100],
    'Salary':[50000,100002000,320000,520000,30000,60000,80000,54000,81000,25000,25000,42000,90000,120000,210000,650000,25000,8952145,2522,62000]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
scaler = StandardScaler()
```

```python
df[['Age', 'Salary']] = scaler.fit_transform(df[['Age', 'Salary']])
```

```python
df
```

```python
scaler = MinMaxScaler()
```

```python
df[['Age', 'Salary']] = scaler.fit_transform(df[['Age', 'Salary']])
```

```python
df
```

```python
# df['Age_Zscore'] = np.abs(stats.zscore(df['Age']) >3)
# df['Salary_Zscore'] = np.abs(stats.zscore(df['Salary']) >3)
```

```python
df['Z-Score_Age'] = zscore(df['Age'])
df['Z-Score_Salary'] = zscore(df['Salary'])
```

```python
outliers_age = df[np.abs(df['Z-Score_Age']) > 3]
outliers_salary = df[np.abs(df['Z-Score_Salary']) > 3]
```
