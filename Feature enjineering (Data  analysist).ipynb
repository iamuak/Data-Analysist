```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
# Task 1 Load And Explore data
data = sns.load_dataset('titanic')
df = pd.DataFrame(data)
df.head(20)
```

```python
df.tail(15)
```

```python
#fill NaN values in columns
df['age'] = df['age'].fillna(0)
df.head(15)
```

```python
# deck columns ae drop because of disturbence
new_df = df.drop(columns = ["deck"], inplace=True)
print(df.head(7))
```

```python
df.isnull().sum()
```

```python
# apply a concatination method
embarked_dummies = pd.get_dummies(df["embarked"], prefix="embarked")
df = pd.concat([df, embarked_dummies], axis=1)
df[["embark_town", "embarked_Q", "embarked_S", "embarked_C"]].head()
```

```python
#use custom mapping is for better work eith numbers
df["sex"] = df["sex"].map({"male": 0, "female": 1})

df["adult_male"] = df["adult_male"].astype(int)

df["alive"] = df["alive"].map({"no": 0, "yes": 1})

who_dummies = pd.get_dummies(df["who"], prefix="who")

df = pd.concat([df, who_dummies], axis=1)
df.head(15)
```

```python
# Task 2 Extracting a new fetures Family Size = sibsp (siblings/spouse) + parch (parents/children)
data = sns.load_dataset('titanic')
df = pd.DataFrame(data)
print(df.head(8))
df['FamilySize'] = df['sibsp'] + df['parch']
df[['FamilySize', 'sibsp', 'parch', 'alone']].head()
```

```python
df[['FamilySize', 'sibsp', 'parch','alone']].head(10)
```

```python
# Task 3 
from sklearn.preprocessing import MinMaxScaler
df = sns.load_dataset('titanic')
df['FamilySize'] = df['sibsp'] + df['parch']
scaler = MinMaxScaler()
df['minmax_age'] = scaler.fit_transform(df[['age']])
df['minmax_fare'] = scaler.fit_transform(df[['fare']])
df['minmax_family_size'] = scaler.fit_transform(df[['FamilySize']])
df.head()
```

```python
#Task 4
from sklearn.preprocessing import PolynomialFeatures
df = sns.load_dataset('titanic')
df['age'] = df['age'].fillna(0)
poly = PolynomialFeatures(degree=2, include_bias=False)
poly_features = poly.fit_transform(df[['age']])
df['age'] = poly_features[:, 0]
df['age_squared'] = poly_features[:, 1]
df[['age', 'age_squared']].head(10)
```

```python
# bonus task
from sklearn.preprocessing import StandardScaler
df = sns.load_dataset('titanic')
df['FamilySize'] = df['sibsp'] + df['parch']
scalar = StandardScaler()
scaled_values = scalar.fit_transform(df[['fare', 'age', 'FamilySize']])
df[['Scaled_fare', 'Scaled_age', 'Scaled_family_size']] = scaled_values
df[['Scaled_fare', 'Scaled_age', 'Scaled_family_size']].head(10)
```

```python
#TAsk 5
df['age_fare_interaction'] = df['age'] * df['fare']
df[['age', 'fare', 'age_fare_interaction']].head(10)
```
